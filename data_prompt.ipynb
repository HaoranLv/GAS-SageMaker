```python
def seq_mix(s1, s2, l1, l2):
    return s1+s2, l1+l2
```

```python
task='asqp'
do_mixup=False
mix_num=1000
```

```python
import numpy as np
train=[]
dev=[]
sent={'观点-负面':'negative', '观点-正面':'positive', '观点-中性':'neutral'}
data_train={}
data_dev={}
data_mix={}
with jsonlines.open('./acdc_data.jsonl') as reader:
    for obj in reader:
        flag=random.random()
        if flag<=0.8:
            data_train[obj['text']]=[]
            if task=='asqp':
                data_train[obj['text']].append((obj['aspect'] if obj['aspect']!='' else 'None', cate[obj['label']], sent[obj['sentiment']], obj['opinion'] if obj['opinion']!='' else 'None'))
            elif task=='aste':
                data_train[obj['text']].append((obj['aspect'] if obj['aspect']!='' else 'None', obj['opinion'] if obj['opinion']!='' else 'None', sent[obj['sentiment']]))
            elif task=='tasd':
                data_train[obj['text']].append((obj['aspect'] if obj['aspect']!='' else 'None', cate[obj['label']], sent[obj['sentiment']]))
        else:
            data_dev[obj['text']]=[]
            if task=='asqp':
                data_dev[obj['text']].append((obj['aspect'] if obj['aspect']!='' else 'None', cate[obj['label']], sent[obj['sentiment']], obj['opinion'] if obj['opinion']!='' else 'None'))
            elif task=='aste':
                data_dev[obj['text']].append((obj['aspect'] if obj['aspect']!='' else 'None', obj['opinion'] if obj['opinion']!='' else 'None', sent[obj['sentiment']]))
            elif task=='tasd':
                data_dev[obj['text']].append((obj['aspect'] if obj['aspect']!='' else 'None', cate[obj['label']], sent[obj['sentiment']]))

for i in range(mix_num):
    s1_idx, s2_idx=np.random.randint(0,len(data_train.keys()),2)
    s,l=seq_mix(list(data_train.keys())[s1_idx], list(data_train.keys())[s2_idx], list(data_train.values())[s1_idx], list(data_train.values())[s2_idx])
    data_mix[s]=l


# for k,v in data_train.items():
#     train.append(k+'####'+str(v))
# for k,v in data_dev.items():
#     dev.append(k+'####'+str(v))

# for k,v in data_mix.items():
#     train.append(k+'####'+str(v))
```
